 ####  [ **1. Model Based Vehicle Tracking for Autonomous Driving in Urban Environments**  ](http://www.roboticsproceedings.org/rss04/p23.pdf) 

##### Overview

Proposes a moving vehicle tracking module that uses dynamic and geometric properties of the tracked vehicle and estimates them using single bayes filter per tracked object.

##### Key points

Model based tracking : What geometric primitives to use and which scenarios to have there applicability.
Dynamic properties: What corresponds to be a dynamic property? Is this the same as descriptor vs feature separation mentioned in the ICP documentation.

Common approaches to tracking: data segmentation and association before doing a filter update stage. Commonly vehicle speed and position are tracked. 

- This paper introduces model based approach which encompasses both geometric and dynamic properties of the tracked vehicle. Concept of anchor point coordinates are introduced to model the same.
- An abstract sensor representation called virtual scan is introduced to model the sensor detection and ease the detection in the point cloud.
- Key assumption: A resonably precise pose of the vehicle is available at all times.
- Though multiple vehicle tracking theoretically can be modelled as a single joint probability distribution, the complexity of the model gets costlier for efficient computation with increased number of vehicles being tracked by the model.
- So disadvantages of joint probability modelling: Complexity of modelling varying number of vehicles, computationally intractable and weak dependencies between vehicles when they are not near each other.
- Assumptions: No two vehicles overlap, Freespace of atleast 1 m around the vehicle.
  

Velocity evolves according to model: _$p(v_t|v_{t-1})$_  
Vehicle moves based on evolved velocity according to dynamic model: _$p(X_t|X_{t-1}, v_t)$_  
Measurements are governed by a measurement model: _$p(Z_t|X_t, G)$_

**Anchor points**  

Tracking of a detected vehicle is considered by assuming a rectangle shape of width W and lenght L in 2D. The tracking is done for centre of gravity of the vehicle, But this assumption is prone to error as the detected CoG is dependent on the angle from which the vehicle is viewed.  
To accomodate for that anchor points are introduced, whose position can change with respect to CoG. 

**Vehicle dynamic model**  
Given velocity v-1 at time step t-1, velocity evolves via addition of random bounded noise based on maximum allowed acceleration _$a_{max}$_ . $\Delta_v$ is sampled uniformly from [-$a_{max}\Delta t$, $a_{max}\Delta t$].

Pose evolves via a a linear model, motion law that is often utilized when exact dynamics of the vehicles are unknown.  
The motion consists of perturbing orientation by $\Delta \theta_1$, then moving forward according to current velocity by _$v_t \Delta t$_ and making final adjustment to orientation by $\theta 2$.  
_$\Delta \theta_1$_ and _$\Delta \theta_2$_ are sampled uniformly from [_$-d\theta_{max} \Delta t$_, _$d\theta_{max} \Delta t$_].

**Sensor data representation**  
Why raw data from 3D Lidar are inefficient?
- Much of data comes from uninteresting topics. e.g. Curb, road, tree tops etc.
- Computational load wasted on processing 3D data, while movement limited to 3D coordinates.
- Additional sensor noise due to constantly changing vantage points due to vehicle being in motion.

Virtual scan:
- Provides a single point of origin for entire data set. Allows constant time lookup for any given point in space.
- An efficient approach to split area around vehicle into occupied, occluded and free.
